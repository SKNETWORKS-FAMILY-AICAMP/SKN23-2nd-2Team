### 2. 저장된 모델 리스트 뽑기

```python
pysql.list_models()
```

### 3. 모델 업로드 (Push)

```python
# 변수 설정
target_model = None  # 학습된 모델 객체
accuracy = 0.0       # 모델 정확도
user_id = 5          # 1정유선 2김도영 3유헌상 4정희영 5정용희

if target_model:
    pysql.push_model(
        model_obj=target_model,
        user_id=user_id,
        accuracy=float(accuracy),
        description="모델 설명 입력"
    )
```
